```python
from fuzzywuzzy import process
import pandas as pd
import numpy as np
from IPython.display import IFrame

df_trending = pd.read_csv("trending_songs.csv")
df_clustered = pd.read_csv("clustered_songs.csv")

choices = [
    "Trending",
    "Acoustic & Instrumental",
    "Dancing",
    "Acoustic",
    "Energetic",
    "Loud & Energetic",
    "Instrumental",
    "Speechy",
    "Lively Music",
    "Energetic Dance",
    "Low Energy",
    "Dance"
]
choices_str = ", ".join(choices)

def play_song(track_id):
    return IFrame(src="https://open.spotify.com/embed/track/"+track_id,
       width="320",
       height="80",
       frameborder="0",
       allowtransparency="true",
       allow="encrypted-media",
      )



def get_song():
    user_input = input(f"what type of music do you want to listen to? The choices are: {choices_str}")
    match, score = process.extractOne(user_input, choices)

    print(f"Playing {match}")
    if match == "Trending":
        single_song = df_trending.sample()
        track_id = single_song["track_id"].values[0]   
        display(play_song(track_id))
    else:
        filtered_df = df_clustered[df_clustered["cluster_name"]== match]
        single_song = filtered_df.sample()        
        track_id = single_song["track_id"].values[0]   
        display(play_song(track_id))


get_song()
```
