```python
import pandas as pd
import numpy as np
from IPython.display import IFrame

df_trending = pd.read_csv("trending_songs.csv")
df_clustered = pd.read_csv("clustered_songs.csv")
```

```python
def play_song(track_id):
    return IFrame(src="https://open.spotify.com/embed/track/"+track_id,
       width="320",
       height="80",
       frameborder="0",
       allowtransparency="true",
       allow="encrypted-media",
      )
```

```python
user_input = input("what type of music do you want to listen to? Trending or 0-11")
if user_input == "trending":
    single_song = df_trending.sample()
    track_id = single_song["track_id"].values[0]   
    display(play_song(track_id))
else:
    filtered_df = df_clustered[df_clustered["cluster"]== int(user_input)]
    single_song = filtered_df.sample()
    print(single_song)
    track_id = single_song["track_id"].values[0]   
    display(play_song(track_id))
```

```python
filtered_df = df_clustered[df_clustered["cluster"] == 5]
single_song = filtered_df.sample()
print(single_song)
track_id = single_song["track_id"].values[0]   
display(play_song(track_id))
```
